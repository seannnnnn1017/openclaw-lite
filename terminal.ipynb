```python
import subprocess

process = subprocess.Popen(
    ["python", WORKPLACE + "test.py"],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
    text=True
)

# 即時讀輸出
for line in process.stdout:
    print("OUT:", line.strip())

for line in process.stderr:
    print("ERR:", line.strip())

process.wait()
```
